**Create a bar chart or histogram to visualize the distribution of a categorical or continuous variable, such as the distribution of ages or genders in a population.**


Dataset: "Wine Quality" (Available from UCI Machine Learning Repository)

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load the data

data = pd.read_csv('https://archive.ics.uci.edu/ml/machine-learning-databases/wine-quality/winequality-red.csv', sep=';')
```

```python
# Explore the data

print("Data Head:\n", data.head())
```

```python
print("\nData Info:\n", data.info())
```

```python
print("\nData Null Values:\n", data.isnull().sum())
```

```python
print("\nData Summary Statistics:\n", data.describe())
```

```python
# Define a function to create a customized plot

def plot_distribution(column, title, xlabel, ylabel, bins=20):

    plt.figure(figsize=(10, 6))

    if data[column].dtype == 'object':
        sns.countplot(x=column, data=data, palette='viridis')
    else:
        sns.histplot(data[column], bins=bins, kde=True, color='skyblue')

    plt.title(title, fontsize=14)
    plt.xlabel(xlabel, fontsize=12)
    plt.ylabel(ylabel, fontsize=12)
    plt.xticks(rotation=45)
    plt.grid(True, alpha=0.3)
    plt.show()
```

```python
# Visualize the distribution of 'quality'

plot_distribution(
    'quality',
    'Distribution of Wine Quality',
    'Quality Score',
    'Frequency'
)
```

```python
# Visualize the distribution of 'alcohol'

plot_distribution(
    'alcohol',
    'Distribution of Alcohol Content',
    'Alcohol Percentage (%)',
    'Frequency',
    bins=30
)
```

```python
# Visualize the distribution of 'pH'

plot_distribution(
    'pH',
    'Distribution of pH Levels',
    'pH Value',
    'Frequency'
)
```
